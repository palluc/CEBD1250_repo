#### Import:  Breed labels data

```python
try:
    conn= mysql.connector.connect(host='localhost', port=3306, user='root',password='pawsware15!',db='pawsware')
    cursor=conn.cursor()
    
    with open('data/breed_labels.csv', 'r') as csvfile:
        breed_csv = csv.reader(csvfile, delimiter=',')
        header = next(breed_csv)
        for row in breed_csv:
            tuple_string = (row[0], row[1], row[2])
            #insertstmt="insert into breed(breed_ID, btype, breedname) values(%s,%s,%s)" %s' ()
            #cursor.execute(insertstmt)
            result = cursor.execute('INSERT INTO breed(breed_ID, btype, breedname)''VALUES(%s, %s, %s)', row)
            conn.commit()
            print ("Record inserted successfully into breed_labels table")

except mysql.connector.Error as error :
    conn.rollback() #rollback if any exception occured
    print("Failed inserting record into breed_labels table {}".format(error))

finally:
    #close database connection.
    if (conn.is_connected()):
        cursor.close()
        conn.close()
        print("MySQL connection is closed")    
```
